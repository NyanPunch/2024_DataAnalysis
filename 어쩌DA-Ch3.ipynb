# 3.1 데이터 필터링과 정렬 테크닉

## 3.1.1 조건식을 활용한 데이터 필터링

```python
import pandas as pd
titanic = pd.read_csv('datasets/train.csv')
titanic.head()
```

```python
# 단일 조건식을 활용하여 데이터 필터링하기
titanic[titanic.Pclass == 3].head()
```

```python
# 다중 조건식을 활용하여 데이터 필터링하기
titanic[(titanic.Pclass == 3) & (titanic.Sex == 'female')].head()
```

```python
# 특정값을 제외한 데이터 추출하기
titanic[(titanic.Pclass == 3) & (titanic.Sex != 'female')].head()
```

## 3.1.2 loc 조건부 필터링

```python
titanic.loc[titanic.Fare > 300,:]
```

```python
titanic.iloc[list(titanic.Fare > 300)]
```

```python
titanic.loc[
    (titanic.Fare > 240) & (titanic.Sex == 'female'), # 다중 조건 입력
    ['Name', 'Age', 'Fare', 'Embarked']               # 특정 컬럼만 선택
]
```

## 3.1.3 특정 값 포함 여부 필터링

```python
# 단일 컬럼에서 특정 값이 포함된 데이터 필터링하기
titanic[titanic.Embarked.isin(['S', 'C'])].head()
```

```python
titanic[~titanic.Embarked.isin(['S', 'C'])].head()
```
